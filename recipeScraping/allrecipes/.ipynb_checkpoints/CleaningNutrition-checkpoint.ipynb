```python
import pandas as pd
df = pd.read_csv("cleanedData3.csv")
```

```python
import json
base_ingredients = dict()
for _, row in df.iterrows():
    ingredients = json.loads(row["ingredients"])
    changed = False
    for ingredient in ingredients:
        try:
            if type(ingredient["base_ingredient"]) == list:
                if ingredient["base_ingredient"][0] == 'noricilantro':
                    ingredient["base_ingredient"] = ["nori", "cilantro", "lime"]
                    changed = True
                for b_i in ingredient["base_ingredient"]:
                    if b_i in base_ingredients:
                        base_ingredients[b_i].append(ingredient["ingredient"])
                    else:
                        base_ingredients[b_i] = [ingredient["ingredient"]]
                    
            else:     
                b_i = ingredient["base_ingredient"]
                if b_i in base_ingredients:
                    base_ingredients[b_i].append(ingredient["ingredient"])
                else:
                    base_ingredients[b_i] = [ingredient["ingredient"]]
        except Exception as e:
            print(ingredient)
            raise e

    if changed:
        row["ingredients"] = json.dumps(ingredients)
```

```python
START = 0
END = 25

for i, key in enumerate(base_ingredients.keys()):
    if i >= END: break
    if i < START: pass
    print(key) 
    print(base_ingredients[key])
```

```python
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        return False


def parse_nutrients(n):
    nutrientDict = dict()
    start_n = n
    nutrients = [
        "Saturated Fat", "Cholesterol", "Sodium", "Total Carbohydrate", "Dietary Fiber", "Total Sugars", "Protein",
        "Vitamin C", "Calcium", "Iron", "Potassium"
                ]

    nutrients = list(filter(lambda x: n.find(x) > -1, nutrients))
    
    n = n.replace("Nutrition FactsServings Per Recipe", "")
    servings = n.split("Calories")[0]
    nutrientDict["servings"] = servings
    n = n.replace(servings + "Calories", "")
    calories = n.split("%")[0]
    nutrientDict["calories"] = calories
    n = n.replace(calories + "% Daily Value *Total ", "")
    
    
    lastNutrient = "Fat" if n.count("fat") > 0 else None
    n = n.replace("Fat", "", 1)
    while len(nutrients) > 0:
        nutrient_positions = [n.find(nut) for nut in nutrients]
        nutrient_position = min(nutrient_positions)
        nutrient = nutrient_positions.index(nutrient_position)
        nutrient_name = nutrients[nutrient]
        if lastNutrient != None:
            nutrient_string = n.split(nutrient_name)[0]
            nutrient_num = None
            nutrient_unit = None
            i = 1
            is_num = is_number(nutrient_string[:i])
            while is_num and i < len(nutrient_string):
                i += 1
                is_num = is_number(nutrient_string[:i])
            i -= 1
            try:
                nutrient_num = float(nutrient_string[:i])
            except Exception as e:
                print(lastNutrient)
                print(nutrient_name)
                print(start_n)
                print(n)
                print(nutrient_string)
                print(i)
                raise e
            nutrient_unit = nutrient_string[i:]
            for i in range(len(nutrient_unit)):
                if is_number(nutrient_unit[i]):
                    nutrient_unit = nutrient_unit[:i]
                    break
            curNutrientDict = dict();
            curNutrientDict["nutrient"] = nutrient_num if nutrient_num != None else nutrient_string
            curNutrientDict["unit"] = nutrient_unit
            nutrientDict[lastNutrient] = curNutrientDict
            
        if nutrient_position > 0:
            n = n[nutrient_position:]

        lastNutrient = nutrient_name
        n = n.replace(nutrient_name, "")
        nutrients.remove(nutrient_name)

    return json.dumps(nutrientDict)
        
    
    
"""
for n in n_facts:
    print(parse_nutrients(n))
"""
```
